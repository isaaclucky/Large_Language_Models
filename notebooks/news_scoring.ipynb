```python
def prepare_prompt(df):
    prompt = []
    for i in range(df.shape[0]):
        p = 'Title: ' + df.iloc[i]['Title'] + '\nDescription: ' + df.iloc[i]['Description'] +'\nBody:'+ df.iloc[i]['Body'][:250] +'\nDomain:'+df.iloc[i]['Domain'] 
        prompt.append(p)
    return prompt
```
